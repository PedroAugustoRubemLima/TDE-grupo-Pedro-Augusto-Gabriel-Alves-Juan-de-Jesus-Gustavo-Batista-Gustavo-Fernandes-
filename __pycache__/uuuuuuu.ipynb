```python
e = [1,2,3,4,5]
r = [10,8,6,4,2]
fig,ax = plt.subplots()
ax.plot(e,r)
ax.set_title("Gráfico de dispersão")
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
```

```python
c = [1,2,3,4,5]
v = [10,8,6,4,2]
fig,ax = plt.subplots()
ax.plot(c,v,
color = "blue",
linestyle ="--",
linewidth=2,
marker="o",
markersize="8",
markerfacecolor="white" )
ax.set_title("Gráfico de dispersão")
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
ax.set_xlim(0,6)
ax.set_ylim(0,12)
ax.grid(True)
```

```python
uh = np.linspace(0,10,100)
hu1 = np.sin(uh)
hu2 = np.cos(uh)
hu3 = np.tan(uh)
fig, axs = plt.subplots(3,1,figsize=(8,10))
axs[0].plot(uh,hu1)
axs[0].set_title("Grafico de Seno")
axs[1].plot(uh,hu2)
axs[1].set_title("Grafico de Cosseno")
axs[2].plot(uh,hu3)
axs[2].set_title("Grafico de Tangente")
```
